```python
# Implementation of Logistic Regression Using Perceptron
# The code implements the process as how the transformation of the region works towards +ve & -ve of the line
# Author: Muhammad Humayun Khan

from sklearn.datasets import make_classification  # using dataset of the sklearn
import numpy as np

X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))  # width is 10 inches while height is 6 inches
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100) # 0 = X-coordinates, 1 = Y-Coordinates, s=100 size of the points
```

```python
def perceptron(X,y):

    X = np.insert(X,0,1,axis=1)     # Add bias term
    weights = np.ones(X.shape[1])   # initialize the weights
    lr = 0.1    # learning rate

    for i in range(1000):   # epoch steps
        j = np.random.randint(0,100)    # total shape is 100, so pick value from 0 to 100
        y_hat = step(np.dot(X[j],weights))    # predicted Y value and send to the step() method defined below
        weights = weights + lr*(y[j]-y_hat)*X[j]  # update label based on actual(yi) and predicated(yi_hat)

    return weights[0],weights[1:]
# step function which returns the placement result
def step(z):
    return 1 if z>0 else 0

intercept_,coef_ = perceptron(X,y)
print(coef_)
print(intercept_)
```

```python
# Now plot the decision boundry in the data

m = -(coef_[0]/coef_[1])        # calculate the slope
b = -(intercept_/coef_[1])      # calculate the y-intercept
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
