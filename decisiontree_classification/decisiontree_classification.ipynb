```python
# Implementation of the Decision Tree Classification
# The dataset used is builtin of the sklearn iris
# Author: Muhammad Humayun Khan

from sklearn.datasets import load_iris                    # import the dataset
from sklearn.model_selection import train_test_split      # for training and testing of data
from sklearn.metrics import accuracy_score                # finding accuracy of the model after training
from sklearn.tree import plot_tree                        # ploting the tree to calculate how tree make decision
from matplotlib.pylab import rcParams                     # to visualize the tree size bigger


iris = load_iris()      # create object of the dataset
X = iris.data           # Features of the iris dataset as X
X
```

```python
# define the target/output as y
y = iris.target
y

# the below output shows three flowers such as
# 0 = setosa
# 1 = versicolor
# 2 = verginica
```

```python
# For training and testing the model, use 4 arrays which store the data
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state = 2)
```

```python
# Import the classifier and train the model
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)
```

```python
# predict the model accuracy score after training
y_pred = clf.predict(X_test)

accuracy_score(y_test,y_pred)     # accuracy of the model
```

```python
# Finding how decision tree make the decision
# lets plot the tree

rcParams['figure.figsize'] = 80,50
plot_tree(clf)



```
